```python
from matplotlib import pyplot as plt

x = range(2,26,2)
y = range(12)

```

```python

# 1. 设置图片大小
# 只需要实例化一次，后续plt 会根据这个配置生成图形
plt.figure(figsize=(10,8),dpi=100)
print(x)
print(y)
plt.plot(x,y)
#plt.savefig("./ret.png")
# # 保存为矢量格式
#plt.savefig("./ret.svg")
# 绘制x轴的刻度
plt.xticks(range(1,25))
plt.yticks(range(min(y),max(y)+1))
#plt.savefig("./ret.png")
plt.show()
```

```python
# demo 120 分钟内温度变化
import random
records = [random.randint(30,35) for i in range(120)]

# matplotlib 默认不支持中文的现实
x_labels = [f"10:{i}" for i in range(0,60,10)]
x_labels +=[f"11:{i}" for i in range(0,60,10)]

from matplotlib import pyplot as plt
plt.figure(figsize=(10,8),dpi=160)
plt.plot(range(1,121),records)
plt.yticks(range(0,40,2))
plt.xticks(range(0,120,10),x_labels,rotation=60) # 第一个参数和第二参数一一对应
plt.xlabel("time")
plt.ylabel("temprature") 
plt.title("temp change chart") 
plt.grid(alpha=0.1) # 添加网格,alpha 设置网格透明度
plt.show()
```
